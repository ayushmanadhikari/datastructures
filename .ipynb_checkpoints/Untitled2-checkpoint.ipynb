```python
class BinarySearchTree:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
    
    def add_child(data):
        if self.data == data:
            return
        else:
            if data<self.data:
                if self.left:
                    self.left.add_child(data)
                else:
                    self.left = BinarySearchTree(data)
            if data>self.data:
                if self.right:
                    self.right.add_child(data)
                else:
                    self.right = BinarySearchTree(data)
    
    def searchTree(self, data):
        if self.data == data:
            return True
        else:
            if data<self.data:
                if self.left:
                    self.left.searchTree(data)
                else:
                    return False
            if data>self.right:
                if self.right:
                    self.right.searchTree(data)
                else:
                    return False
    
    def inOrderTraversal(self):
        elements = []
        if self.left:
            elements = elements + self.left.inOrderTraversal()
        elements.append(self.data)
        if self.right:
            elements = elements + self.right.inOrderTraversal()
        
        return elements
    #pre and post order are same except when the data is appended first in preorder and
    # data is appended last at post order
    
    def findMin(self):
        elements = self.inOrderTraversal()
        min = elements[0]
        return min
    
    def findMax(self):
        if self.right is None:
            return self.data
        else:
            self.right.findMax()
    
    def findSum(self):
        elements = self.inOrderTraversal()
        sum1 = 0
        for e in elements:
            sum1 = sum1 + int(e)
        return sum1 
    
    
        
            
        
        
                    
                    
                    
                    
                    
```
