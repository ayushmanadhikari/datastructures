```python
new_df = pd.read_excel('formatteddata2.xlsx', dtype = {'Item Details': 'string', 'Qty': 'float', 'Unit': 'string', 'Vch/Bill No': 'string', 'Particulars': 'string', 'Date': 'string' })
```

```python
new_df
```

```python
# concatenating Item Details column entries grouping by date, vch, particulars
temp_df_item = new_df.groupby(['Date', 'Vch/Bill No', 'Particulars'])['Item Details'].apply(', '.join).reset_index()
temp_df_item
```

```python
# concatenating Unit also removing duplicates in that cell
temp_df_unit = new_df.groupby(['Date', 'Vch/Bill No', 'Particulars'])['Unit'].apply(', '.join).reset_index()
temp_df_unit['Unit'] = temp_df_unit['Unit'].str.split(', ').apply(set).str.join(', ')
temp_df_unit
```

```python
# getting the resultant sum of QTY from grouped by date, vch, particulars
temp_df_qty = new_df.groupby(['Date', 'Vch/Bill No', 'Particulars'], as_index=False).sum('Qty')
temp_df_qty
```

### Joining above 3 DF - Qty, Unit and Item Details into a single DF


```python
tempdf1 = temp_df_item.merge(temp_df_unit, on=['Date', 'Vch/Bill No', 'Particulars'])
tempdf2 = tempdf1.merge(temp_df_qty, on=['Date', 'Vch/Bill No', 'Particulars'])
tempdf2
```

```python
#tempdf2.to_excel('output_file.xlsx', index=False)
if os.path.exists('final_output.xlsx'):
    os.remove('final_output.xlsx')
    print('old file deleted and new one created!')
tempdf2.to_excel('final_output.xlsx', index=False)
```
