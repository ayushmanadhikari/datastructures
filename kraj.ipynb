```python
import pandas as pd
```

```python
df1 = pd.read_excel('kraj/report.xlsx', 'Sheet1')
df2 = pd.read_excel('kraj/report.xlsx', 'Sheet2')
```

```python
df1 = df1.astype({'Bank Account Number': 'string', 'Member Name': 'string'})
df1_list = df1[['Member Name', 'Bank Account Number']]
df1_list.head(10)
```

```python
df2 = df2.rename(columns={'Unnamed: 0': 's.n', 'Unnamed: 1': 'Particular', 'Unnamed: 2': 'Name', 'Unnamed: 3': 'amount1', 'Unnamed: 4': 'amount2', 'Unnamed: 5': 'a/c no.'})
df2 = df2[1:]
df2.astype({'Name': 'string', 'a/c no.': 'string'})
```

### Assuming that sheet 2 contains the correct account name and number

```python
df2_list = df2[['Name', 'a/c no.']]
df2_list = df2_list.astype({'Name': 'string', 'a/c no.': 'string'})
df2_list.dtypes
```
